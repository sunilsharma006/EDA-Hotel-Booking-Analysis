# Below is the description of the Columns/ Variables in our dataset
**Hotel:** Type of hotel (City hotel or Resort hotel)

**is_cancelled:** If the booking was cancelled (1) or not (0)

**lead_time:** Number of days before the actual arrival of the guests

**arrival_date_year:** Year of arrival date (2015-2017)

**arrival_date_month:** Month of arrival date

**arrival_date_week_number:** Week number of year for arrival date

**arrival_date_day_of_month:** Day of arrival date

**stays_in_weekend_nights:** Number of weekend nights(Saturday or Sunday) spent at the hotel by the guests.

**stays_in_week_nights:** Number of week nights(Monday to Friday) spent at the hotel by the guests.

**adults:** Number of adults among the guests

**children:** Number of children

**babies:** Number of babies

**meal:** Type of meal booked (BB-Breakfast, HB- Half Board, FB- Full Board, SC- Self Catering)

**country:** country of the guests

**market_segment:** Designation of market segment(Aviation, Complementary, Corporate, Direct, Groups, Inline TA, Offline TO/TA) Where TO and TA stands for Tour Operator and Travel Agent respectively.

**distribution_channel:** Name of booking distribution channel (Corporate, Direct, GDS, TA/TO)

**is_repeated_guest:** If the booking was from a repeated guest(1) or not(0)

**previous_cancellation:** Number of previous bookings that were cancelled by the customer prior to the current booking

**previous_bookings_not_cancelled:** Number of previous bookins not cancelled by the customer prior to the current booking

**reserved_room_type:** Code from room type reserved

**assigned_room_type:** Code of room type assigned

**booking_changes:** Number of changes made to the booking

**deposit_type:** Type of deposite made by the guest

**agent:** ID of travel agent who made the booking

**comapny:** ID of the company that made the booking

**days_in_waiting_list:** Number of the days the booking was in the waiting list

**customer_type:** Type of customer, assuming one of four categories

**adr:** Average daily rate

**required_car_parking_spaces:** Number of car parking spaces required bt the customer

**total_of_special_requesrs**: Number of special requests made by the customer

**reservation_statuse:** Reservation status(Canceled, check-out or no-show)

**reservation_status_date:** Date at which the last reservation status was updated

# **Removal of Duplicate data**

```python
# Rows containing duplicate data
duplicate_rows_data = hotel_booking_df[hotel_booking_df.duplicated()]
print("Number of duplicate rows", duplicate_rows_data.shape)
```

```python
# Dropping duplicate Values
hotel_booking_df.drop_duplicates(inplace = True)
hotel_booking_df.shape
```

```python
# Updated dataset after dropping duplicate values
hotel_booking_df.info()
```

# **Finding Missing Values**

```python
# Getting summary of missing values present in the dataset
for column in hotel_booking_df:
  if hotel_booking_df[column].isnull().any():
    print('Column {0} has {1} missing values'. format(column, hotel_booking_df [column]. isnull().sum(), round(hotel_booking_df[column].isnull().sum())))
```

```python
# The 'company' column has significant missing values so we may drop this column as this will not be useful for us.
hotel_booking_df = hotel_booking_df.drop(['company'], axis = 1)
```

**Null values may be handelled in two ways.**
1. These may be filled with some appropriate data
2. The rows with null values may be deleted, but this will reduce the size of dataset.

Here we are applying the first approach.

```python
# Filling null values
#Filling null values of agent with 0
hotel_booking_df ['agent'].fillna(0,inplace = True)
#Filling null values of children with mean
hotel_booking_df ['children'].fillna(0,inplace = True)
#Filling null values of country with unknown
hotel_booking_df ['country'].fillna('unknown', inplace = True)
```

**Conversion of datatypes**

```python
# Now let's convert datatypes of columns'children' and 'agent' from float to int
hotel_booking_df[['children', 'agent']] = hotel_booking_df[['children', 'agent']].astype(int)
```

```python
# Changing datatype of column 'reservation_status_date' to datetime
hotel_booking_df['reservation_status_date'] = pd.to_datetime(hotel_booking_df['reservation_status_date'])
format = '%y-%m-%d'
```

**Addition of new Columns for simplifying analysis**

```python
# Let's group some of the columns which can be fruitful in analysis as grouped element
hotel_booking_df ['total_stay'] = hotel_booking_df['stays_in_weekend_nights'] + hotel_booking_df['stays_in_week_nights']
hotel_booking_df ['total_guests'] = hotel_booking_df['adults'] + hotel_booking_df['children'] + hotel_booking_df['babies']
```

```python
hotel_booking_df.describe() # Here we are finding little presence of outliers. ( As in adr column highest value is 5400, it may be an outlier, we need to check it.)
```

```python
# Try to find the presence of outlier.
hotel_booking_df['adr'].plot(kind='box')
```

As shown there is presence of outliers in dataset. Which is responsible for the deviation of mean value to a large extent.

```python
# Removing the above outlier,  for adr column we are selecting values below 5000 only.
hotel_booking_df= hotel_booking_df[hotel_booking_df['adr']<5000] 

hotel_booking_df['adr'].plot(kind='box')
```

```python
# Checking the updated information about the dataset
hotel_booking_df.info()
```

# **Data Analysis and Visualization**

```python
# Cancellation percentage of booking
canceled_perc= hotel_booking_df['is_canceled'].value_counts(normalize= True)
print(canceled_perc)
```

```python
# Plotting the bar chart to show the above variation.
plt.figure(figsize=(8,5))
plt.title("Reservation Status")
plt.bar(['Not Cancelled','Cancelled'], hotel_booking_df['is_canceled'].value_counts(), edgecolor='maroon', width = 0.6)
plt.show()
```

```python
# Now we need to check that for bookings and cancellation, how much part is for 'Resort hotel and for City hotel'.
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel', hue='is_canceled', data=hotel_booking_df, palette='Oranges')
legend_labels,_=ax1.get_legend_handles_labels()
# ax1.legend(bbox_to_anchor(1,1))
plt.title("Reservation Status in different Hotels", size=25)
plt.xlabel('hotel')
plt.ylabel('Number of reservation')
plt.show()
```

```python
# Plotting a comparative chart for month-wise booking and cancellation data.
hotel_booking_df['month']= hotel_booking_df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x ='month', hue= 'is_canceled',  data = hotel_booking_df, palette= 'bright')
legend_labels = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor= (1,1))
plt.title('Reservation Status per month', size= 20)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.legend(['Not_cancelled', 'cancelled'])
plt.show()
```
